```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
df.head()
```

```python
df = df.set_index(df['timestamp'])
del df['timestamp']
df
```

```python
df.index
df['day']=df.index.day
df['month'] = df.index.month
df['year'] = df.index.year
df['dayofyear'] = df.index.dayofyear
df
```

```python
prices_per_year = df.groupby('year')[['open','high', 'low', 'close', 'volume']].mean().reset_index()
prices_per_year
```

```python
def findPrice(x):
    diff = x['close']-x['open']
    return pd.Series([x['year'],diff,x['open'],x['close']],index=['year','diff','open','close'])
mean_price_year = prices_per_year.apply(findPrice, axis=1)
mean_price_year
```

```python
prices_per_day_of_year = df.groupby(['year', 'dayofyear'])[['open','high','low','close', 'volume']].mean().reset_index()
prices_per_day_of_year
```

```python
fig = plt.figure(figsize=(15,5))
plt.plot(prices_per_year['year'],prices_per_year['close'],color='green')
plt.bar(prices_per_year['year'], prices_per_year['volume']*100, width=0.3, color='blue')
plt.xlabel("Years")
plt.ylabel("Price")
plt.title("Bitcoin price per year")
```

```python
years = prices_per_day_of_year['year'].unique()
years
```

```python
fig = plt.figure(figsize=(15,10))

for year in years:
    year_data = prices_per_day_of_year[prices_per_day_of_year['year']==year]
    fig=plt.subplots(figsize=(15,5))
    plt.plot(year_data['dayofyear'],year_data['close'],color='blue')
    plt.title(year)
    plt.xlabel("Days")
    plt.ylabel("Price")
```

```python
prices_by_month = df.groupby(['year','month'])[['open','close','high','low','volume']].mean().reset_index()
prices_by_month
```

```python
for year in years:
    year_data = prices_by_month[prices_by_month['year']==year]
    fig = plt.subplots(figsize=(7,5))
    plt.plot(year_data['month'],year_data['close'],color='red')
    plt.title(year)
    plt.xlabel("Months")
    plt.ylabel("Price")
```

```python
for year in years:
    year_data = prices_by_month[prices_by_month['year']==year]
    fig = plt.figure(figsize=(7,5))
    up = year_data[year_data['close']>=year_data['open']]
    down = year_data[year_data['close']<year_data['open']]
    col1="red"
    col2="green"
    width1=0.2
    width2 = 0.03
    #print(up)
    #print(down)
    #Plotting up prices of the stock
    plt.bar(up['month'], (up['close']-up['open'])*10000, width1, bottom = up['open'], color=col2)
    plt.bar(up['month'], (up['high']-up['close'])*10, width2, bottom=up['close'], color=col2)
    plt.bar(up['month'], (up['low']-up['open'])*10, width2, bottom=up['open'], color=col2)
    
    #Plotting down prices of the stock
    plt.bar(down['month'], (down['close']-down['open'])*10000, width1, bottom=down['open'], color=col1)
    plt.bar(down['month'], (down['high']-down['open'])*10, width2, bottom=down['open'], color=col1)
    plt.bar(down['month'], (down['low']-down['close'])*10, width2, bottom=down['close'], color=col1)
    
    
    plt.show()
```
